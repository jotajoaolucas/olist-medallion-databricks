```python
%sql
CREATE CATALOG IF NOT EXISTS olist_project;

CREATE SCHEMA IF NOT EXISTS olist_project.bronze;
CREATE SCHEMA IF NOT EXISTS olist_project.silver;
CREATE SCHEMA IF NOT EXISTS olist_project.gold;

-- Volume para receber os arquivos brutos do Kaggle
CREATE VOLUME IF NOT EXISTS olist_project.bronze.raw_files;
```

```python
import os

with open("/Volumes/olist_project/bronze/raw_files/kaggle_token.txt") as f:
    token = f.read().strip()

os.environ["KAGGLE_API_TOKEN"] = token

import kaggle
kaggle.api.authenticate()
print("✅ Autenticado com sucesso")
```

```python
kaggle.api.dataset_download_files(
    "olistbr/brazilian-ecommerce",
    path="/tmp/olist",
    unzip=True
)

import shutil
for f in os.listdir("/tmp/olist"):
    shutil.copy(f"/tmp/olist/{f}", f"/Volumes/olist_project/bronze/raw_files/{f}")

print(os.listdir("/Volumes/olist_project/bronze/raw_files"))
```
